```
class Personne():
  def __init__(self,genes,hair,eyes,poids,taille,age,name):
    self.hair = hair
    self.eyes = eyes
    self.poids = poids
    self.taille = taille
    self.genes = genes
    self.age = age
    self.name = name

  
  def get_IMC(self): 
     return self.poids / (self.taille*self.taille) 
gene_S = {'sequence':'ACCATGGTCATGCGTAGCATGT',
           'genes':{
               'genes1':(0,4),
               'genes2':(4,7),
               'genes3':(12,18),
               'genes4':(7,12),
               'genes5':(18,22)
               
                }
          }
gene_R = {'sequence':'CTTAGAACTTGGGTGACTGAC',
           'genes':{
               'genes1':(0,5),
               'genes2':(5,9),
               'genes3':(12,17),
               'genes4':(9,12),
               'genes5':(17,22)
               
                }
          }
  
```

```
sarah = Personne(gene_S,hair='chatin',eyes='marron',poids=59,taille=1.64,age=22,name="sarat")
raouf = Personne(gene_R,hair='chatin',eyes='noisette',poids=65,taille=1.66,age=23,name="titiiich")
```

```
%time
from dna_features_viewer import GraphicFeature,GraphicRecord
def print_dna(genes):
  sequence_genes = genes['sequence']
  features_gene = []
  for gene_ in genes['genes'].values():
    features_gene.append(GraphicFeature(start=gene_[0],end=gene_[1],strand=+1, color='#ffcccc'))
  record = GraphicRecord(sequence=sequence_genes,features=features_gene)
  ax,_= record.plot(figure_width=20)
  record.plot_sequence(ax)
  record.plot_translation(ax,(1,len(sequence_genes)), fontdict={'weight':'bold'})
print_dna(sarah.genes)
print_dna(raouf.genes)
```

```
%time
list_genes_S = []
list_genes_R = []
for gene_ in sarah.genes['genes'].values():
  list_genes_S.append(sarah.genes['sequence'][gene_[0]:gene_[1]])
for gene_ in raouf.genes['genes'].values():
  list_genes_R.append(raouf.genes['sequence'][gene_[0]:gene_[1]])
print(f'le ADN de {sarah.name} est:',list_genes_S)
print(f'le ADN de {raouf.name} est:',list_genes_R)
```

```
%time
nbr_genes_commun = 0
for gene_S in list_genes_S:
  if gene_S in list_genes_R:
    print('gene en commun!',gene_S)
    nbr_genes_commun = nbr_genes_commun +1
print(f'le porcentage de la similarité entre {sarah.name} et {raouf.name} est: ', nbr_genes_commun*5 /100,"%")
```
